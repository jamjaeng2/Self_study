# 11729 하노이 탑 이동 순서

```python
# 하노이의 탑
def hanoi(n, from_pos, to_pos, aux_pos):
    if n == 1:
        print('a', from_pos, "->", to_pos)
        return
    print('d', n-1, from_pos, aux_pos, to_pos)
    hanoi(n - 1, from_pos, aux_pos, to_pos)
    
    print('b',from_pos, "->", to_pos)
    print()
    
    print('c', n-1, aux_pos, to_pos,from_pos)
    hanoi(n-1, aux_pos, to_pos, from_pos)

#print("n = 3")
hanoi(4, 1, 3, 2)
```

# 2447 별 찍기

```python
n = int(input())
star_array = [[0 for _ in range(n)] for _ in range(n)]
#print(star_map)





def star_map(n):
    if n == 3:
        star_array[0][0:3] = star_array[2][0:3] = [1] * 3
        star_array[1][0:3] = [1, 0, 1]
        return
    
    a = n//3
    star_map(a)
    
    for i in range(3):
        for j in range(3):
            
            if i == j == 1:
                continue
            for k in range(a):
                star_array[a*i + k][a*j:a*(j+1)] = star_array[k][:a]
star_map(n)

for i in star_array:
    for j in i:
        if j:
            print("*", end='')
        else:
            print(" ", end='')
    print()
            
```
